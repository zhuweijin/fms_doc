```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [80, 60, 70, 90, 85]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

ax.fill(categories, values, color='skyblue', alpha=0.25)
ax.plot(categories, values, color='skyblue', linewidth=2)
ax.fill(categories, values, color='skyblue', alpha=0.25)
ax.set_ylim(0, 100)
ax.set_yticks(np.arange(0, 100, 20))

plt.show()
```

```python
import pyecharts.options as op
from pyecharts.charts import Radar

# ,6
v1 = [[100, 100, 50, 50, 50]]
v2 = [[50, 50, 75, 100, 100]]
v3 = [[100, 75, 50, 75, 75]]


# ,
x_schema = [
        {"name": "",  "color": 'black', "font_size": 18},
        {"name": "",  "color": 'black', "font_size": 18},
        {"name": "",  "color": 'black', "font_size": 18},
        {"name": "",  "color": 'black', "font_size": 18},
        {"name": "",  "color": 'black', "font_size": 18}
    ]

# 
radar_x = Radar()
radar_x.add_schema(x_schema)
radar_x.add('', v1, color='red').set_colors(['red'])
radar_x.add('', v2, color='green').set_colors(['green'])
radar_x.add('', v3, color='orange').set_colors(['orange'])

radar_x.set_global_opts(
#         title_opts=op.TitleOpts(pos_right="center"),
        legend_opts=op.LegendOpts(legend_icon="roundRect", align="left", pos_left='7%',
                                  pos_bottom='14%', orient='vertical')
    )

radar_x
radar_x.render("cloud_type.html")
```
